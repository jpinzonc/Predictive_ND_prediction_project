```python
import pandas as pd
import numpy as np
```

```python
sales = pd.read_csv('p2-2010-pawdacity-monthly-sales-p2-2010-pawdacity-monthly-sales.csv')
sales.shape
```

```python
compt = pd.read_csv('p2-partially-parsed-wy-web-scrape.csv')
compt.head()
```

```python
demog = pd.read_csv('p2-wy-demographic-data.csv')
demog.head()
```

```python
sales.loc[:,'Total Pawdacity Sales'] = sales[['January',	'February'	,'March'	,'April',	'May'	,'June'	,'July',	'August',	'September',	'October'	,'November'	,'December']].sum(1)
sales
```

```python
comp1 = compt["City|County"].str.split("|", expand = True).rename(columns = {0: 'CITY'}).drop(1,1).merge(compt, left_index = True, right_index = True)[['CITY', '2010 Census']]
comp1.loc[:,'CITY'] = comp1.CITY.str.replace(r' \?|\?', '')
comp1 = comp1[comp1.CITY.isin(sales.CITY.unique())]
comp1.loc[:,'2010 Census Population'] = comp1['2010 Census'].str.replace('<td>|</td>|\,', "").astype(np.float)
comp1= comp1.drop(['2010 Census'],1)
comp1
```

```python
final = demog[demog.City.isin(sales.CITY.unique())].merge(sales[['CITY', 'Total Pawdacity Sales']], left_on = 'City', right_on = 'CITY').merge(comp1, on= 'CITY').drop(['County', 'CITY'],1)
```

```python
ordered_col = ['City',
'2010 Census Population',
'Total Pawdacity Sales',
'Households with Under 18',
'Land Area',
'Population Density',
'Total Families']
np.round(final[ordered_col].mean(),2)
```

```python
data_out = final.reset_index(drop = True)
out_df = final.drop('City', 1)
outliers_all = np.array([], dtype='int64')
for column in data_out.drop('City', 1).columns:
    Q1 = np.percentile(out_df[column], 25)
    Q3 = np.percentile(out_df[column], 75)
    step = (Q3 - Q1) * 2
    out_list_feat=list(out_df[~((out_df[column] >= Q1 - step) & (out_df[column] <= Q3 + step))].index)
    outliers_all = np.append(outliers_all, out_list_feat)

from collections import Counter
out_count = Counter(outliers_all)
out_customer = []
for customer, count in out_count.items():
    if count > 1:
        out_customer.append(customer.astype(int))
outliers  = list(sorted(out_customer))

print("%d outliers found in more than one feature will be remove" % len(outliers))
data_no_out = data_out.drop(data_out.index[outliers])
print(data_out.shape, data_no_out.shape)
```
